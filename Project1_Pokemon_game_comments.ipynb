```python
def Pokemon_game():
    #import matplotlib, random and colorama libraries
    from matplotlib import pyplot as plt
    from matplotlib import image as mpimg
    import random
    import colorama
    from colorama import Fore
    from time import sleep
    print('Lets Duel !!!!!!!')
    sleep(3.0)
    #define list of pokemons
    
    pokemon_list=['pikachu','evee','charmander','squirtle','onix','bulbasaur','pidgey','pidgeot','blastoise','raichu','charizard','vinasaur']
    
    #define type of each pokemon
    pokemon_types={'pikachu':'electric','raichu':'electric','blastoise':'water','charizard':'fire','vinasaur':'leaf','pidgey':'flying','evee':'ground','charmander':'fire','squirtle':'water','onix':'rock','bulbasaur':'leaf','pidgeot':'flying','zubat':'bug','butterfree':'bug'}
    
    #define each pokemon's moves
    pokemon_moves={'pikachu':['tackle','thunder','quick attack','thunder wave'],'evee':['tackle','quick attack','growl','bite'],'squirtle':['bubble','tackle','water gun','growl'],'bulbasaur':['vine whip','tackle','growl','bite'],'charmander':['ember','tackle','quick attack','growl']
                  ,'pidgeot':['gust','wing attack','tackle','fly'],'onix':['rock slide','rock tomb','tail whip','tackle'],'charizard':['flamethrower','ember','tackle','wing attack'],'pidgey':['gust','tackle','growl','quick attack'],'vinasaur':['vine whip','razor leaf','tackle','bite'],
                   'blastoise':['water blast','water gun','bubble','bite'],'raichu':['thunder bolt','thunder','thunder wave','tackle']}
    #define what move types are super effective on each pokemon type
    effective_moves={'electric':['water','flying'],'fire':['leaf','bug'],'water':['fire','rock'],'leaf':['water','rock'],'rock':['bug','flying','ground','electric'],'flying':['bug','leaf'],'bug':['leaf'],'ground':''}
    not_effective_moves={'electric':['rock','electric'],'fire':['water','rock','fire'],'water':['leaf','thunder','water'],'leaf':['fire','bug','leaf'],'rock':['rock'],'flying':['rock','flying'],'bug':['bug','rock','flying'],'ground':['rock']}
    #define power of each move
    move_powers={'gust':15,'tail whip':1,'wing attack':30,'fly':20,'rock slide':20,'rock tomb':25,'tackle':10,'thunder':20,'quick attack':15,'growl':1,'thunder wave':30,'bite':15,'bubble':20,'water gun':30,'ember':20,'vine whip':20,'thunder bolt':30,'water blast':30,'razor leaf':35,'flamethrower':35}
    #define type of each move
    move_types={'tackle':'ground','quick attack':'ground','bite':'ground','thunder':'electric','thunder wave':'electric','bubble':'water','water gun':'water','vine whip':'leaf','razor leaf':'leaf','ember':'fire','rock slide':'rock','rock tomb':'rock','wing attack':'flying','thunder bolt':'electric','gust':'flying','fly':'flying','tail whip':'ground','growl':'ground','flamethrower':'fire','water blast':'water'}
    #define each possible pokemon evolution
    pokemon_evolutions={'pikachu':'raichu','squirtle':'blastoise','pidgey':'pidgeot','charmander':'charizard','bulbasaur':'vinasaur'}
    
    #ask user 1 to choose pokemon from list
    user1_pokemon=input(f"choose from {pokemon_list} ")
    #show picture of selected pokemon using mpimg.read
    image = mpimg.imread(f"Documents/{user1_pokemon}.png")
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    sleep(2.0)
    
    #ask user 2 to choose pokemon from list
    user2_pokemon=input(f"choose from {pokemon_list} ")
    image = mpimg.imread(f"Documents/{user2_pokemon}.png")
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    sleep(2.0)
    #set initial health points to 100
    user1_hp=100
    user2_hp=100
    
    #enter while loop that continues executing as long as both pokemon's health points are greater than 0
    while user1_hp >0 and user2_hp>0:
        #ask user 1 to choose his/her pokemon's move
        user1_chosen_move=input(Fore.BLACK+ f"choose {user1_pokemon}'s move from {pokemon_moves[user1_pokemon]} ")
        #in case move not on list or misspelt, ask again
        while user1_chosen_move not in  pokemon_moves[user1_pokemon]:
            user1_chosen_move=input(Fore.BLACK+ f"choose  {user1_pokemon}'s move again from {pokemon_moves[user1_pokemon]} ")
        #check if move chosen is super effective,not effective or neutral on opponent. Check if critical hit is possible
        if pokemon_types[user2_pokemon] in effective_moves[move_types[user1_chosen_move]]:
            eff=10
            print(Fore.BLACK+ 'Its super effective')
        elif pokemon_types[user2_pokemon] in not_effective_moves[move_types[user1_chosen_move]]:
            eff=-10
            print(Fore.BLACK+ 'Its not very effective')
        elif random.choice(range(1,101)) in range(10,101,10):
            eff=30
            print(Fore.GREEN + 'Critical hit')
        else:
            eff=0
        #reduce opponent's health points accordingly
        user2_hp=user2_hp-move_powers[user1_chosen_move]-eff
        #if opponent's hp drops to 0 or lower, user 1 wins, break from loop
        if user2_hp <=0:
            print(Fore.BLACK+ f"{user2_pokemon} has 0 hp left")
            print (Fore.BLACK+ 'user1 wins')
            break
        #display remaining health points
        print(Fore.BLACK+ f"{user2_pokemon} has {user2_hp} hp left")
        
        #ask user 2 to choose his/her's pokemon's move
        user2_chosen_move=input(Fore.BLACK+ f"choose {user2_pokemon}'s move  from {pokemon_moves[user2_pokemon]} ")
        #ask again if name is misspelt or not in list
        while user2_chosen_move not in  pokemon_moves[user2_pokemon]:
            user2_chosen_move=input(Fore.BLACK+ f"choose  {user2_pokemon}'s move again from {pokemon_moves[user2_pokemon]} ")
        #check if move chosen is super effective,not effective or neutral on opponent. Check if critical hit is possible
        if pokemon_types[user1_pokemon] in effective_moves[move_types[user2_chosen_move]]:
            eff=10
            print(Fore.BLACK+ 'Its super effective')
        elif pokemon_types[user1_pokemon] in not_effective_moves[move_types[user2_chosen_move]]:
            eff=-10
            print(Fore.BLACK+ 'Its not very effective')
        elif random.choice(range(1,101)) in range(10,101,10):
            eff=30
            print(Fore.GREEN+ 'Critical hit')
        else:
            eff=0 
        #reduce opponent's health points accordingly
        user1_hp=user1_hp-move_powers[user2_chosen_move]-eff
         #if opponent's hp drops to 0 or lower, user 2 wins, break from loop
        if user1_hp <=0:
            print(Fore.BLACK+ f"{user1_pokemon} has 0 hp left")
            print (Fore.BLACK+ 'user2 wins')
            break
        #diplay remaining hp 
        print(Fore.BLACK+ f"{user1_pokemon} has {user1_hp} hp left")
        #check if pokemon will involve via conditons of reaching 70 hp or less and can evolve
        if user1_hp<=70 and user1_hp>0 and user1_pokemon in ['pikachu','charmander','squirtle','bulbasaur','pidgey']:
            rand_numb=[23,45,23,0,12,34,23,23,23,25,95,23,0]
            if random.choice(rand_numb) in [23,0]:
                old_poke1=user1_pokemon
                user1_hp+=20
                print(Fore.RED + "....")
                sleep(5.0)
                print(Fore.RED + 'Hold on!')
                sleep(5.0)
                print(Fore.RED + f"{user1_pokemon} is evolving!!!")
                user1_pokemon=pokemon_evolutions[user1_pokemon]
                sleep(10.0)
                print(Fore.RED + f"{old_poke1} evolved to {user1_pokemon}!!!!!!")
                sleep(5.0)
                image = mpimg.imread(f"Documents/{user1_pokemon}.png")
                plt.imshow(image)
                plt.xticks([])
                plt.yticks([])
                plt.show()
                print(Fore.RED + f"{user1_pokemon} now has {user1_hp} hp left")
                sleep(2.0)
        if user2_hp<=70 and user2_hp>0 and user2_pokemon in ['pikachu','charmander','squirtle','bulbasaur','pidgey']:
            rand_numb=[23,45,23,0,12,34,23,23,23,25,95,23,0]
            if random.choice(rand_numb) in [23,0]:
                old_poke2=user2_pokemon
                user2_hp+=20
                print(Fore.BLUE+ "....")
                sleep(5.0)
                print(Fore.BLUE+ 'Hold on!')
                sleep(5.0)
                print(Fore.BLUE+ f"{user2_pokemon} is evolving!!!")
                user2_pokemon=pokemon_evolutions[user2_pokemon]
                sleep(10.0)
                print(Fore.BLUE+ f"{old_poke2} evolved to {user2_pokemon} !!!!!!!")
                sleep(5.0)
                image = mpimg.imread(f"Documents/{user2_pokemon}.png")
                plt.imshow(image)
                plt.xticks([])
                plt.yticks([])
                plt.show()
                print(Fore.BLUE+ f"{user2_pokemon} now has {user2_hp} hp left")
                sleep(2.0)
  
Pokemon_game()
```
